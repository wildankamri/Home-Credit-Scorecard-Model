# Project Background
Banyak orang berjuang untuk mendapatkan pinjaman karena tidak mencukupi atau tidak mempunyai riwayat kredit. Dan sayangnya, orang-orang ini sering dimanfaatkan oleh pemberi pinjaman yang tidak dapat dipercaya. Home Credit berupaya untuk memperluas inklusi keuangan bagi orang-orang yang tidak memiliki rekening bank dengan memberikan pengalaman pinjaman yang aman. Untuk memastikan orang-orang yang kurang terlayani ini memiliki pengalaman pinjaman yang positif, Home Credit menggunakan berbagai metode statistik dan machine learning untuk membuat prediksi kemampuan pembayaran klien. Melakukan hal itu akan memastikan bahwa klien yang mampu membayar tidak ditolak dan bahwa pinjaman diberikan dengan pokok, jatuh tempo, dan kalender pembayaran yang akan memberdayakan klien untuk menjadi sukses.

# Goal
Membangun model untuk memprediksi kemampuan setiap pemohon untuk membayar kembali pinjaman (credit score)


# Objective
Membuat model prediktif untuk menentukan klien potensial

```python
def miss_numerical(df_train2):
    
    numerical_features = df_train2.select_dtypes(exclude = ["object"] ).columns
    for f in numerical_features:
        df_train2[f] = df_train2[f].fillna(df_train2[f].median())
    return df_train2

def miss_categorical(df_train2):
    
    categorical_features = df_train2.select_dtypes(include = ["object"]).columns
    for f in categorical_features:
        df_train2[f] = df_train2[f].fillna(df_train2[f].mode()[0])
    return df_train2

def transform_feature(df_train2):
    df_train2 = miss_numerical(df_train2)
    df_train2 = miss_categorical(df_train2)
    return df_train2

df_train2 = transform_feature(df_train2)
pd.set_option("max_columns", None)
df_train2.head()
```

```python
# Membuat 2 dataframe berbedda dari kelas majority dan minority  
df_majority = df_train2[(df_train2['TARGET']==0)] 
df_minority = df_train2[(df_train2['TARGET']==1)] 

# Upsample kelas minority 
df_minority_upsampled = resample(df_minority, 
                                 replace=True,    
                                 n_samples= 282686,
                                 random_state=42)  

# Menggabungkan kelas majority dengan upsampled kelas minority 
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
fig = plt.figure(figsize = (12,7))
df_upsampled['TARGET'].value_counts(normalize=True).plot(kind='bar', color= ['blue','darkgrey'], alpha = 0.9, rot=0)
plt.title('Distribusi dari kemampuan klien membayar\n', fontsize=14)
plt.ylabel('Persentase klien\n', fontsize=12)
plt.xlabel('\nPayment Difficulty Status', fontsize=12)
plt.show()
```
